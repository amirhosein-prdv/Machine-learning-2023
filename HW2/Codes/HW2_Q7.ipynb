```python
import matplotlib.pyplot as plt
import numpy as np
```

## Create Data


```python
N = 1000
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
np.shape(X)
```

## Plot Histogram

```python
plt.hist(X,100 , color='blue',edgecolor='black');
plt.xlabel('value'); plt.ylabel('quantitiy')
plt.title('Histogram plot of generated data');
```

### Gaussian Kernel Function


```python
Gaussian_kernel = lambda x : (1/(np.sqrt(2*np.pi)))*np.exp((x**2)/-2)
```

## Plot density for different h

```python
x_d = np.linspace(-4, 8, 2000)
for h in [0.5, 0.1,1,10]:
    density = sum(Gaussian_kernel((xi - x_d)/h) for xi in X)/N
    plt.figure()
    plt.plot(x_d, density, '-k')
    plt.fill_between(x_d, density, alpha=0.5 , color='blue', label=f'Window size (h) = {h}')
    plt.xlabel('x'); plt.ylabel('value');
    plt.title('distribution of samples')
    plt.legend(loc='best')
```
